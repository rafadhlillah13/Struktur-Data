Mengacu pada soal Jurnal nomor 7, buatlah Queue ADT menggunakan metode Stack sebagai variabel instance yang akan menghasilkan output sebagai berikut:

Output: \\

1 1 \\

```python
class Stack:
  def __init__ (self):
    self._data = []
  
  def __len__(self):
    return len(self._data)
  
  def isempty(self):
    return len(self._data) == 0

  @property
  def view(self):
    return self._data

  def push(self, item):
    self._data.append(item)

  def pop(self):
    if self.isempty():
      raise Exception("stack kosong")
    return self._data.pop()

  def top(self):
    if self.isempty():
      raise Exception("stack kosong")
    return self._data[-1]
```

```python
class StackWithQueue:
  def __init__(self):
    self._data = Stack()
    self._size = 0

  def __len__(self):
    return self._size

  def isempty(self):
    return self._size == 0
  
  @property
  def view(self):
    return self._data.view
    
 
  def first(self):
    return self._data.view[0]
  
  def last(self):
    return self._data.view[-1]
   
  def enqueue(self,item):
    self._data.push(item)
    self._size += 1
    for i in range (self._size -1):
      self._data.push(self._data.pop())

  def dequeue(self):
    if self.isempty():
      raise Exception('Stack kosong')
    else:
      self._size -= 1
      return self._data.pop()
```

```python
SWQ = StackWithQueue()

for i in range(10):
  SWQ.enqueue(i)

SWQ.view

for out in SWQ.view :
  print(out, out)
```

##Nomor 2
Mengacu pada soal Jurnal nomor 4, buatlah program Deque ADT yang dapat
menggambarkan proses step by step ketika penambahan elemen secara add_first dan
add_last! Hint: Buat sebuah array pada self._data yang berisikan None.
Outputnya seperti berikut:

image.png

```python
class DequeADT:
  def __init__(self):
    self._data = [None]*10
    self._size = 0
  
  def __len__(self):
    return self._size

  def isempty(self):
    return len(self._data) == 0
  
  @property
  def view(self):
    return self._data
 
  def first(self):
    return self._data[0]
  
  def last(self):
    return self._data[-1]
  
  def add_first(self,item):
    if (None in self._data) == False:
      self._data = self._data + ([None]*10)
    for i in range(len(self._data), 0, -1):
      if self._data[i-1] == None:
        self._data[i-1] = item
        break
    self._size += 1
    return self._data
  
  def add_last(self,item):
    if (None in self._data) == False:
      self._data = self._data + ([None]*10)
    for i in range(len(self._data)):
      if self._data[i] == None:
        self._data[i] = item
        break
    self._size += 1
    return self._data

  def delete_first(self):
    if self.isempty():
      raise Exception('Stack kosong')
    else:
      self._size -= 1
      return self._data.pop(0)

  def delete_last(self):
    if self.isempty():
      raise Exception('queue kosong')
    else:
      self._size -= 1
      return self._data.pop()
```

```python
ADT = DequeADT()

print("Adding Last")
for i in range(10):
  print(i ,ADT.add_last(i))

print("Adding first")
for i in range(20, 10, -1):
  print(i ,ADT.add_first(i))
```

##Nomor 3
Buatlah sebuah fungsi yang dapat menampilkan hasil permutasi dari sebuah array yang berisikan kombinasi bilangan menggunakan modul Stack yang telah dibuat di kelas!

Output: \\
[5, 4, 3, 2, 1] \\
[5, 4, 3, 1, 2] \\
[5, 4, 2, 3, 1] \\
[5, 4, 2, 1, 3] \\
[5, 4, 1, 3, 2] \\
[5, 4, 1, 2, 3], dst

```python
class Stack:
  def __init__(self):
    self._data = []
    
  def __len__(self):
    return len(self._data)

  def isempty(self):
    return len(self._data) == 0
  
  @property
  def view(self):
    return self._data

  def push(self,item):
    self._data.append(item)

  def pop(self):
    if self.isempty():
      raise Exception('Stack kosong')
    return self._data.pop()

  def top(self):
    if self.isempty():
      raise Exception('Stack kosong')
    return self._data[-1]
```

```python
def kombinasi(arr):
  if len(arr) == 0:
    return []
  if len(arr) == 1:
    return [arr]
  r = []
  for i in range(len(arr)):
    a = arr[i]
    b = arr[:i]+arr[i+1:]
    for k in kombinasi(b):
      r.append([a]+k)
  return r
```

```python
kombinasi_bilangan = Stack()
for i in range(5, 0, -1):
  kombinasi_bilangan.push(i)
for i in kombinasi(kombinasi_bilangan.view):
  print(i)
```
