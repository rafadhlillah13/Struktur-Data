```python
class LinkedQueue:
  def __init__(self, df):
    self._df = df
    r = 0
    for i in self._df: #membuat array
      if r == 0:
        self._nama = i
      elif r == 1:
        self._sex = i
      elif r == 2:
        self._age = i
      r += 1
  def __len__(self):
    return len(self._df[self._nama])
  def is_empty(self): 
    return len(self._df[self._nama]) == 0
  def first(self):
    if self.is_empty():
      raise Exception('Queue is empty')
    nama = self._df[self._nama][0]
    gender = self._df[self._sex][0]
    umur = self._df[self._age][0]
    view = [nama, gender, umur]
    return view
  def last(self):
    if self.is_empty():
      raise Exception('Queue is empty')
    nama = self._df[self._nama][len(self._df[self._nama])-1]
    gender = self._df[self._sex][len(self._df[self._nama])-1]
    umur = self._df[self._age][len(self._df[self._nama])-1]
    view = [nama, gender, umur]
    return view
  def view(self):
    return self._df
  def add_first(self, item): #fungsi untuk menambahkan data di depan
    nama = self._df[self._nama].tolist()#konversi ke list
    sex = self._df[self._sex].tolist()
    age = self._df[self._age].tolist()
    nama.insert(0, item[0])#menambah data (.insert => menambah data pada posisi array tertentu)
    sex.insert(0, item[1])
    age.insert(0, item[2])
    llist = {self._nama : nama, self._sex : sex, self._age : age}
    self._df = pd.DataFrame(llist)
  def add_last(self, item): #fungsi untuk menambahkan data dibelakang
    nama = self._df[self._nama].tolist()#konversi ke list
    sex = self._df[self._sex].tolist()
    age = self._df[self._age].tolist()
    nama.append(item[0])#menambah data (.append => menambah data array diakhir/dibelakang)
    sex.append(item[1])
    age.append(item[2])
    llist = {self._nama : nama, self._sex : sex, self._age : age}
    self._df = pd.DataFrame(llist)
  def delete_first(self): #fungsi untuk menghapus data di depan
    nama = self._df[self._nama].tolist()#konversi ke list
    sex = self._df[self._sex].tolist()
    age = self._df[self._age].tolist()
    nama.pop(0)#hapus data (.pop)
    sex.pop(0)
    age.pop(0)
    llist = {self._nama : nama, self._sex : sex, self._age : age}
    self._df = pd.DataFrame(llist)
  def delete_last(self):
    nama = self._df[self._nama].tolist()#konversi ke list
    sex = self._df[self._sex].tolist()
    age = self._df[self._age].tolist()
    nama.pop()#hapus data (.pop)
    sex.pop()
    age.pop()
    llist = {self._nama : nama, self._sex : sex, self._age : age}
    self._data = pd.DataFrame(llist)
```
